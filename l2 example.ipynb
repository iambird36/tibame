# while loop

```python
i = 1
while i < 6:
    print(i)
    i += 1
```

```python
i = 0
num_sum = 0
while i <= 100:
    num_sum = num_sum + i
    i = i + 1
print('1+2+......+100 = {}'.format(num_sum))
```

```python
i = 0
num_sum = 0
while i <= 100:
    if i % 2 == 1:
        i = i + 1
        continue
    num_sum = num_sum + i
    i = i + 1
print('2+4+6......+100 = {}'.format(num_sum))
```

```python
i = 0
num_sum = 0
while True:
    if i > 500:
        break
    num_sum = num_sum + i
    i = i + 1
print('1+2+......+100 = {}'.format(num_sum))
```

# nested loop

```python
for i in range(5):
    for j in range(5):
        print('i: {}, j: {}'.format(i, j))
```

```python
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adj:
    for y in fruits:
        print(x, y)
```

```python
n=5;
for i in range(n):
    for j in range(i):
        print ('* ', end="")
    print('')

for i in range(n,0,-1):
    for j in range(i):
        print('* ', end="")
    print('')
```

```python
'''

pip install jieba

精確模式 ：將句子最精確地切開，叫適合文本分析。
寫法:words = jieba.cut(content, cut_all=False)

全模式：把句子中所有的可以成詞的詞語都掃描出來, 速度快。
寫法:words = jieba.cut(content, cut_all=True)

搜索引擎模式：在精確模式的基礎上對長詞再次切分，提高召回率，適合用於搜尋引擎分詞。 寫法:jieba.cut_for_search(Content)


'''

import jieba

sentence = "足球運動需要大家一起來推廣，歡迎加入我們的行列！"
print("輸入： {}".format(sentence))
words1 = jieba.cut(sentence, cut_all=False)
words2 = jieba.cut(sentence, cut_all=True)
words3 = jieba.cut_for_search(sentence)

print("精確模式：", end=' ')
for word in words1:
    print(word+'/', end='')

print('')

print("全模式：", end=' ')
for word in words2:
    print(word+'/', end='')

print('')

print("搜索引擎模式：", end=' ')
for word in words3:
    print(word+'/', end='')
```

```python
'''
pip install wordcloud
ref: https://github.com/amueller/word_cloud
'''

%matplotlib inline
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 

f = open("english_new.txt", "r")


comment_words = f.read().lower().strip()
stopwords = set(STOPWORDS) 
  
#print(STOPWORDS)
print(comment_words)  
  
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.savefig("123.png",bbox_inches = "tight")  
plt.show()

```

```python
def show_list_items(*items):
    for i in items:
        print(i)
my_list = ['apple', 'banana', 'orange']
show_list_items(my_list)
```

```python
def show_dict_items(**items):
    for i in items:
        print(i)
my_dict = {'isaac':80, 'andy':90, 'amy':30}
show_list_items(my_dict)
```

```python
def list_statistic(u_list):
    num_sum = sum(u_list)
    num_avg = num_sum / len(u_list) 
    return num_sum, num_avg

my_list = [100, 20, -30, -50] 
print(list_statistic(my_list))
```
