# Read Speech.txt file, use jieba to do word segmentation and do word frequency on each word

```python
import jieba
import csv
ret = open("./speech.txt", "r").read().strip()

result={}
word = jieba.cut(ret, cut_all=False)
print(word)
for c in word:
    if c in result:
        result[c] = result[c] +1
    else:
        result[c] = 1
print(result)
with open("count.csv","w", encoding='utf-8-sig') as fd:
    w=csv.writer(fd)
    w.writerow(result.keys())
    w.writerow(result.values())
```

# Use wordcloud to show the word frequency in the file "speech.txt"

```python
%matplotlib inline
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt 
import jieba


## write your code here

ret = open("./speech.txt", "r").read().strip()

word = jieba.cut(ret, cut_all=False)
word = " ".join(word)    
stopwords = set(STOPWORDS) 

 
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', font_path='simhei.ttf',
                stopwords = stopwords,
                min_font_size = 10).generate(word) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show()

## show chinese in wordcloud: use WordCloud(font_path='simhei.ttf', ......)
```
