link: https://www.ptt.cc/bbs/movie/M.1586525586.A.891.html

```python
# -*- coding: utf-8 -*-
import requests
from bs4 import BeautifulSoup
import re

url = 'https://www.ptt.cc/bbs/movie/M.1586525586.A.891.html'

## write your code here
response = requests.get(url)
soup =BeautifulSoup(response.text, "html.parser")
content = soup.find("div", {"id":"main-content"})
meta = content.find_all("div", class_="article-metaline")
for m in meta:
    m.extract()
right_meta = content.find_all("div", class_="article-metaline-right")
for s in right_meta:
    s.extract()
pushes = content.find_all("div", class_="push")
for s in pushes:
    s.extract()
tail = content.find_all("span", class_="f2")
for s in tail:
    s.extract()
print(content.text.replace('--', ''))

```

```python
%matplotlib inline
from wordcloud import WordCloud,STOPWORDS
import jieba
import matplotlib.pyplot as plt 

## write your code here
ret = content.text.replace('--', '')
word = jieba.cut(ret, cut_all=False)
word = " ".join(word)   
stopwords = set(STOPWORDS) 

 
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', font_path='simhei.ttf',
                stopwords = stopwords,
                min_font_size = 10).generate(word) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show()
```
